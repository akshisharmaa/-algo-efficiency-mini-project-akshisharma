```python
import time
import matplotlib.pyplot as plt
import sys
import sys
import threading

def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)


def get_max_recursion_depth(n):
    return n

n_values = list(range(1, 30))
execution_times = []
space_usages = []

for n in n_values:
    start = time.perf_counter()
    fibonacci_recursive(n)
    end = time.perf_counter()
    execution_times.append(end - start)

    approx_frame_size = 100
    space_usages.append(get_max_recursion_depth(n) * approx_frame_size)

plt.figure(figsize=(10, 6))

fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('n (input size)')
ax1.set_ylabel('Time (seconds)', color=color)
ax1.plot(n_values, execution_times, marker='o', color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_yscale('log')

ax2 = ax1.twinx()

color = 'tab:green'
ax2.set_ylabel('Space (bytes)', color=color)
ax2.plot(n_values, space_usages, marker='x', color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Time vs Space Complexity of Recursive Fibonacci')
fig.tight_layout()
plt.grid(True)
plt.show()

print("\n--- Recursive Fibonacci Analysis ---")
print("Input/Output:")
print("  Input: n (integer, position in Fibonacci sequence)")
print("  Output: nth Fibonacci number")
print("\nTime Complexity:")
print("  Best Case: O(1) when n = 0 or 1")
print("  Worst Case: O(2^n) due to repeated recursive calls")
print("  Average Case: O(2^n) (same as worst case)")
print("\nSpace Complexity:")
print("  O(n) because of recursion stack depth (approximation based on recursion depth)")
```

```python
import time
import matplotlib.pyplot as plt
import sys

def fibonacci_dp(n):
    if n <= 1:
        return n
    dp = [0, 1]
    for i in range(2, n + 1):
        dp.append(dp[i - 1] + dp[i - 2])
    return dp[n], dp

n_values = [100, 500, 1000, 2000, 5000, 10000]
execution_times = []
space_usage = []

for n in n_values:
    start = time.perf_counter()
    result, dp = fibonacci_dp(n)
    end = time.perf_counter()
    execution_times.append(end - start)

    size_dp = sys.getsizeof(dp) + sum(sys.getsizeof(x) for x in dp)
    space_usage.append(size_dp)

plt.figure(figsize=(10, 6))

plt.plot(n_values, execution_times, marker='o', color='red', label='Time (seconds)')
plt.plot(n_values, space_usage, marker='x', color='purple', label='Space (bytes)')
plt.title('Time vs Space Complexity of DP Fibonacci')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

print("Execution Times (seconds):", execution_times)
print("Space Usage (bytes):", space_usage)
```

```python
import time
import random
import sys
import matplotlib.pyplot as plt

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

sizes = [100, 500, 1000, 2000, 3000, 4000, 5000]
time_taken = []
space_used = []

random.seed(42)

for size in sizes:
    arr = [random.randint(1, 10000) for _ in range(size)]


    total_size = sys.getsizeof(arr) + sum(sys.getsizeof(x) for x in arr)
    space_used.append(total_size)

    start = time.perf_counter()
    bubble_sort(arr.copy())
    end = time.perf_counter()

    time_taken.append(end - start)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(sizes, time_taken, marker='o', color='blue', label='Execution Time')
ax1.set_xlabel('Array Size (n)')
ax1.set_ylabel('Execution Time (seconds)', color='blue')
ax1.tick_param_
```

```python
import time
import matplotlib.pyplot as plt
import sys
import random


def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result


input_sizes = [10, 50, 100, 200, 400, 800]
execution_times = []
space_usage = []

for size in input_sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]

    start = time.time()
    merge_sort(arr)
    end = time.time()
    execution_times.append(end - start)

    space_usage.append(size * sys.getsizeof(arr[0]))

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='green', label='Space (bytes)')
plt.title('Time vs Space Complexity of Merge Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()


print("\n--- Merge Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")
print("\nTime Complexity:")
print("  Best Case: O(n log n)")
print("  Average Case: O(n log n)")
print("  Worst Case: O(n log n)")
print("\nSpace Complexity:")
print("  O(n) due to temporary arrays used in merging")
```

```python
import time
import matplotlib.pyplot as plt
import sys
import random

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

input_sizes = [10, 50, 100, 200, 400, 800]
execution_times = []
space_usage = []

random.seed(42)

for size in input_sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]

    start = time.perf_counter()
    insertion_sort(arr.copy())
    end = time.perf_counter()
    execution_times.append(end - start)

    total_space = sys.getsizeof(arr) + sum(sys.getsizeof(x) for x in arr)
    space_usage.append(total_space)

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='green', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='orange', label='Space (bytes)')
plt.title('Time vs Space Complexity of Insertion Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

print("\n--- Insertion Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")
print("\nTime Complexity:")
print("  Best Case: O(n) when array is already sorted")
print("  Average Case: O(n^2)")
print("  Worst Case: O(n^2) when array is reverse sorted")
print("\nSpace Complexity:")
print("  O(1) → In-place sorting (no extra space needed)")

```

```python
import time
import matplotlib.pyplot as plt
import sys
import random


def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)


input_sizes = [10, 50, 140, 256, 440, 870]
execution_times = []
space_usage = []

for size in input_sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]

    start = time.time()
    quick_sort(arr)
    end = time.time()
    execution_times.append(end - start)

    space_usage.append(sys.getsizeof(arr) + (size * sys.getsizeof(arr[0])))


plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='red', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='purple', label='Space (bytes)')
plt.title('Time vs Space Complexity of Quick Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()


print("\n--- Quick Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")
print("\nTime Complexity:")
print("  Best Case: O(n log n) when pivot divides array evenly")
print("  Average Case: O(n log n)")
print("  Worst Case: O(n^2) when pivot is smallest/largest element repeatedly")
print("\nSpace Complexity:")
print("  O(log n) for recursion stack in best/average cases")
print("  O(n) in worst case recursion depth")
```

```python
import time
import matplotlib.pyplot as plt
import sys
import random


def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

input_sizes = [10, 50, 100, 200, 400, 800]
execution_times = []
space_usage = []

random.seed(42)

for size in input_sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]

    start = time.perf_counter()
    selection_sort(arr.copy())
    end = time.perf_counter()
    execution_times.append(end - start)

    total_space = sys.getsizeof(arr) + sum(sys.getsizeof(x) for x in arr)
    space_usage.append(total_space)

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='red', label='Space (bytes)')
plt.title('Time vs Space Complexity of Selection Sort')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

print("\n--- Selection Sort Analysis ---")
print("Input/Output:")
print("  Input: Array of n integers")
print("  Output: Sorted array")

print("\nTime Complexity:")
print("  Best Case: O(n^2) (no early termination possible)")
print("  Average Case: O(n^2)")
print("  Worst Case: O(n^2)")

print("\nSpace Complexity:")
print("  O(1) → In-place sorting (no extra space needed)")
```

```python
import time
import matplotlib.pyplot as plt
import sys
import random




def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

input_sizes = [10, 50, 100, 200, 400, 800, 1600]
execution_times = []
space_usage = []

for size in input_sizes:
    arr = sorted([random.randint(1, 10000) for _ in range(size)])
    target = random.choice(arr)

    start = time.time()
    binary_search(arr, target)
    end = time.time()
    execution_times.append(end - start)

    space_usage.append(sys.getsizeof(arr) + (size * sys.getsizeof(arr[0])))

plt.figure(figsize=(8, 5))
plt.plot(input_sizes, execution_times, marker='o', color='blue', label='Time (seconds)')
plt.plot(input_sizes, space_usage, marker='o', color='red', label='Space (bytes)')
plt.title('Time vs Space Complexity of Binary Search')
plt.xlabel('Input Size (n)')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

print("\n--- Binary Search Analysis ---")
print("Input/Output:")
print("  Input: Sorted array of n integers and a target element")
print("  Output: Index of target element (or -1 if not found)")

print("\nTime Complexity:")
print("  Best Case: O(1) → Target found at middle on first try")
print("  Average Case: O(log n)")
print("  Worst Case: O(log n) → Target not found or at extreme end")

print("\nSpace Complexity:")
print("  Iterative: O(1) → Constant extra space")
print("  Recursive: O(log n) → Due to recursion stack")
```
